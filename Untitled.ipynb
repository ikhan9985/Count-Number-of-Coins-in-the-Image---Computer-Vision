```python
import cv2

def is_inside_contour(contour, contours):
    """
    Check if a contour is fully contained inside another contour.
    
    Parameters:
        contour (numpy.ndarray): A contour as returned by cv2.findContours().
        contours (List[numpy.ndarray]): A list of contours to check against.
        
    Returns:
        bool: True if the contour is fully inside another contour, False otherwise.
    """
    for other_contour in contours:
        if contour is not other_contour and cv2.pointPolygonTest(other_contour, tuple(map(int, contour[0][0])), False) >= 0:
            # If any point of the contour is outside the other contour, return False
            return False
    # If the contour is fully inside another contour, return True
    return True

# Load the image
img = cv2.imread('e4.jpg')

# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur to reduce noise
blur = cv2.GaussianBlur(gray, (5, 5), 0)

# Apply Canny edge detection
edges = cv2.Canny(blur, 50, 200)

# Find contours in the edges image
contours, hierarchy = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Filter out inner contours
outer_contours = []
for contour in contours:
    if is_inside_contour(contour, outer_contours):
        outer_contours.append(contour)

# Draw the outer contours on the original image in green color
cv2.drawContours(img, outer_contours, -1, (0, 255, 0), 2)

# Print the number of detected coins
print("Number of coins detected:", len(outer_contours))

# Display the image with detected contours
cv2.imshow("Coins", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
